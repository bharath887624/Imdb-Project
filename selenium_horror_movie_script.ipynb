```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()  #Webdriver as object

driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=horror") #Open the webpage

time.sleep(3) #Wait for page to load

#Function to click more
#//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button/span/span ~ XPATH of more()
def click_more():
  try:
    more_button = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button/span/span')

    #Scroll to click more
    ActionChains(driver).move_to_element(more_button).perform()

    #click more button
    more_button.click()

    #wait for page to load
    time.sleep(3)
    return True
  except Exception as e: #except block
    print("Error clicking 'More' button:", e)
    return False
  
#Click more button until no longer there
while click_more():
    print("Clicked 'More' button")

#No more button
print('No more button')
```

```python
import pandas as pd
#list to store data
# Movie Name
# Genre is action
# Ratings
# Voting Counts
# Duration
movie_names = []
ratings = []
votings = []
durations = []

#loop the movie item on page
#Xpath : //*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li
#declare the element
movie_items = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')#'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')

for movie_item in movie_items:
  try:
    movie_name = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[1]/a/h3').text #Xpath for movie name

    rating = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[1]').text #Xpath for rating

    voting = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[2]').text #Xpath for voting

    duration = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[2]/span[2]').text #Xpath for duration

    movie_names.append(movie_name)
    ratings.append(rating)
    votings.append(voting)
    durations.append(duration)

  except Exception as e:
    #print(f"Error extracting data for a movie: {e}") #if element not found skip and print message
    continue

#dataframe
df = pd.DataFrame({
    'Movie_name': movie_names,
    'Rating': ratings,
    'Votes': votings,
    'Duration': durations
})

df.to_csv('imdb_horror_movies_2024.csv', index=False)
driver.quit()
df.describe()
```
